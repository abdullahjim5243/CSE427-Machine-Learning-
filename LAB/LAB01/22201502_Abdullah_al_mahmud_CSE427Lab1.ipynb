```python
import numpy as np
```

```python
a = np.random.randint(1, 100, 20)
b = np.random.randint(1, 100, 20)
print("Array a:", a)
print("Array b:", b)
```

```python
a = a.reshape(5, 4)
b = b.reshape(5, 4)
print("\nMatrix a:", a)
print("Matrix b:", b)
```

```python
a_scal = a * 2
b_scal = b * 2
print(a_scal)
print(b_scal)
```

```python
b_mod = b.T
c = np.dot(a,b_mod)
print(c)
```

```python
print(c[:, 1:3])


print(c[::])
```

```python
largest_element = c.max()
print(largest_element)

flattened_index = c.argmax()
print(flattened_index)
```

```python
c_flat = c.flatten()
print(c_flat)
```

```python
import matplotlib.pyplot as plt
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Temperature vs Humidity (Dhaka)")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.plot(study_hours, marks, color='blue', marker='o', linestyle='-', linewidth=2)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.barh(subjects, marks, color='orange')
plt.title("Horizontal Bar Graph of Subject Scores")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
```

```python
colors = ['red', 'green', 'blue', 'purple', 'cyan']
plt.bar(subjects, marks, color=colors)
plt.title("Vertical Bar Graph of Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_study = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_subjects = [89, 90, 45, 78, 99]

# 1.
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')

# 2.
plt.subplot(2, 2, 2)
plt.plot(study_hours, marks_study, color='blue', marker='o')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')

# 3.
plt.subplot(2, 2, 3)
plt.barh(subjects, marks_subjects, color='orange')
plt.title('Scores in Subjects')
plt.xlabel('Marks')
plt.ylabel('Subjects')

# 4.
colors = ['red', 'green', 'blue', 'purple', 'cyan']
plt.subplot(2, 2, 4)
plt.bar(subjects, marks_subjects, color=colors)
plt.title('Subject Scores')
plt.xlabel('Subjects')
plt.ylabel('Marks')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
